## Nearest Neighbor Algorithm

The main idea of this algorithm is quite naive:

> Start at any city; at each step extend the tour by moving from the previous city to its nearest neighbor that has not yet been visited. Repeat until the starting city is reached.

The distance between all cities are calculated based on Euclidean distance. The time copmplexity of such an algorithm is O(n<sup>2</sup>).
The implementation of this approach is illustated in the cell below:

```python
def nn_tsp(cities):
    """Start the tour at the first city; at each step extend the tour 
    by moving from the previous city to the nearest neighboring city, C,
    that has not yet been visited."""
    start = first(cities)
    tour = [start]
    unvisited = set(cities - {start})
    while unvisited:
        C = nearest_neighbor(tour[-1], unvisited)
        tour.append(C)
        unvisited.remove(C)
    return tour

def nearest_neighbor(A, cities):
    "Find the city in cities that is nearest to city A."
    return min(cities, key=lambda c: distance(c, A))

def shortest_tour(tours): 
    "Choose the tour with the minimum tour length."
    return min(tours, key=tour_length)

def first(collection):
    "Start iterating over collection, and return the first element."
    return next(iter(collection))


def distance(A, B):
    """Calculate distance between two points"""  
    return math.sqrt((B[0]-A[0])**2 + (B[1]-A[1])**2)
```

Now, let's include the plot functions in order to get a more intuitive idea of how this algorithm works.

```python
def plot_tour(tour): 
    "Plot the cities as circles and the tour as lines between them."
    start = tour[0]
    plt.figure(figsize=(10, 10))
    plot_lines(list(tour) + [tour[0]])
    plot_lines([start], 'rs') # Mark the start city with a red square
    
def plot_lines(points, style='bo-'):
    "Plot lines to connect a series of points."    
    plt.plot([p[0] for p in points], [p[1] for p in points], style)
    plt.axis('scaled'); plt.axis('off')
    
def tour_length(tour):
    #print(type(tour))
    "The total of distances between each pair of consecutive cities in the tour."
    return sum(distance(tour[i], tour[i-1]) 
               for i in range(len(tour)))
```

```python
def plot_tsp(algorithm, cities):
    "Apply a TSP algorithm to cities, plot the resulting tour, and print information."
    # Find the solution and time how long it takes
    t0 = perf_counter()
    tour = algorithm(cities)
    t1 = perf_counter()
    assert valid_tour(tour, cities)
    plot_tour(tour); plt.show()
    print("{} city tour with length {:.1f} in {:.3f} secs for {}"
          .format(len(tour), tour_length(tour), t1 - t0, algorithm.__name__))
    
def valid_tour(tour, cities):
    "Is tour a valid tour for these cities?"
    return set(tour) == set(cities) and len(tour) == len(cities)
```

Since near-optimal solutions to TSP instances demand much less computational effort, we will use the whole instance of cities.

```python
cities = frozenset(zip(df['X'],df['Y']))
plot_tsp(nn_tsp, cities)
```

We notice that the solution for an instance of 395 cities has been calculated in less than 0.1 second! But since our results generate some highly redundant moves, let's try to improve this approach.

## Repeated Nearest Neighbor Algorithm

Nearest Neighbour algorithm can be improved by following this approach:

> For each of the cities, run the nearest neighbor algorithm with that city as the starting point, and choose the resulting tour with the shortest total distance.

Since we are going to apply this algorithm for each city, we will increase the run time n times longer.

```python
def repeated_nn_tsp(cities):
    "Repeat the nn_tsp algorithm starting from each city; return the shortest tour."
    return shortest_tour(nn_tsp(cities, start) 
                         for start in cities)
```

We make some modifications in the algorithm introduced earlier:

```python
def nn_tsp(cities, start=None):
    """Start the tour at the first city; at each step extend the tour 
    by moving from the previous city to its nearest neighbor 
    that has not yet been visited."""
    if start is None: start = first(cities)
    tour = [start]
    unvisited = set(cities - {start})
    while unvisited:
        C = nearest_neighbor(tour[-1], unvisited)
        tour.append(C)
        unvisited.remove(C)
    return tour
```

```python

plot_tsp(nn_tsp, cities)
plot_tsp(repeated_nn_tsp, cities)
```

Algortihm | Time Complexity | Length | Computation Time
--- | --- | --- | --- |
Nearest Neighbour | O(n<sup>2</sup>) | 1637.5 | 0.077 sec
Repeated Nearest Neighbour | O(n<sup>2</sup>) | 1547.0 | 27.105 sec 

> We notice that the shortest tour by Repeated Nearest Neighbor Algorithm has been improved in comparison to that of Nearest Neighbor Algorithm. However with respect to their lengths. Nevertheless, it demands more computation time.

## Sampled Repeated Nearest Neighbor Algorithm

We can modify the Repeated Nearest Neighbor Algorithm so that it can be **executed for specific number of cities**.

```python
def repeated_nn_tsp(cities, repetitions=None):
    "Repeat the nn_tsp algorithm starting from specified number of cities; return the shortest tour."
    return shortest_tour(nn_tsp(cities, start) 
                         for start in sample(cities, repetitions))

def sample(population, k, seed=42):
    "Return a list of k elements sampled from population. Set random.seed with seed."
    if k is None or k > len(population): 
        return population
    random.seed(len(population) * k * seed)
    return random.sample(population, k)
```

We test the Sampled Repeated Nearest Neighbor Algorithm for two cases:
* start from 10 different cities, and
* start from 100 different cities

```python
def repeat_10_nn_tsp(cities): return repeated_nn_tsp(cities, 10)
def repeat_100_nn_tsp(cities): return repeated_nn_tsp(cities, 100)
```

And now we can compare them with the initial Nearest Neighbor Algorithm:

```python
plot_tsp(nn_tsp, cities)
plot_tsp(repeat_10_nn_tsp, cities)
plot_tsp(repeat_100_nn_tsp, cities)
```

## Benchmarking for TSP Algorithms

It's now time to compare the computation time of the previous three approaches.

```python
@functools.lru_cache(None)
def benchmark(function, cities):
    "Run function on all the inputs; return pair of (average_time_taken, results)."
    t0           = perf_counter()
    results      = function(cities)
    t1           = perf_counter()
    execution_time = t1 - t0
    return (execution_time, results)
```

```python
def benchmarks(tsp_algorithms, cities):
    "Print benchmark statistics for each of the algorithms."    
    for tsp in tsp_algorithms:
        time, results = benchmark(tsp, cities)
        lengths = tour_length(results) 
        print("Algorithm: {:>15}, executed in: {:7.8f} secs for: {:3.0f} cities, with total tour length: {:5.3f}"
              .format(tsp.__name__, time, len(cities), lengths))
```

```python
algorithms = [nn_tsp, repeat_10_nn_tsp, repeat_100_nn_tsp]

benchmarks(algorithms, frozenset(zip(df['X'],df['Y'])))
```

Algortihm | Time Complexity | Length | Computation Time
--- | --- | --- | --- |
Nearest Neighbour | O(n<sup>2</sup>) | 1637.5 | 0.077 sec
Repeated Nearest Neighbour for 10 cities | O(n<sup>2</sup>) | 1549.4 | 0.616 sec 
Repeated Nearest Neighbour for 100 cities | O(n<sup>2</sup>) | 1547.0 | 6.118 sec 

We notice that we managed to improve the shortest tour with the repeated Nearest Neighbor Algorithm. However, despite the fact that we increased the number of starting cities from 10 to 100, the shortest tour hasn't significantly improved. 

## Greedy Algorithm

Another known heuristic algorithm is the Greedy Algorithm: it's mentioned as Greedy since it greedily adds to the tour the edge that is shortest. Its main steps are presented below:

> Maintain a set of segments; intially, each city defines its own 1-city segment. Find the shortest possible edge that connects two endpoints of two different segments, and join those segments with that edge. Repeat until we form a segment that tours all the cities.

The time copmplexity of this algorithm is **O(n<sup>2</sup>logn)**.
The implementation of this approach is illustated in the cell below:

```python
def greedy_tsp(cities):
    """Go through edges, shortest first. Use edge to join segments if possible."""
    endpoints = {c: [c] for c in cities} # A dict of {endpoint: segment}
    for (A, B) in shortest_edges_first(cities):
        if A in endpoints and B in endpoints and endpoints[A] != endpoints[B]:
            new_segment = join_endpoints(endpoints, A, B)
            if len(new_segment) == len(cities):
                return new_segment
```

```python
def shortest_edges_first(cities):
    "Return all edges between distinct cities, sorted shortest first."
    edges = [(A, B) for A in cities for B in cities 
                    if id(A) < id(B)]
    return sorted(edges, key=lambda edge: distance(*edge))
```

```python
def join_endpoints(endpoints, A, B):
    "Join B's segment onto the end of A's and return the segment. Maintain endpoints dict."
    Asegment, Bsegment = endpoints[A], endpoints[B]
    if Asegment[-1] is not A: Asegment.reverse()
    if Bsegment[0] is not B: Bsegment.reverse()
    Asegment.extend(Bsegment)
    del endpoints[A], endpoints[B] # A and B are no longer endpoints
    endpoints[Asegment[0]] = endpoints[Asegment[-1]] = Asegment
    return Asegment
```

```python
plot_tsp(greedy_tsp, cities)
```

## Altered Greedy Algorithm

At this point, we will try to **improve greedy algorithm** by **swapping segments**.

```python
def altered_greedy_tsp(cities):
    "Run greedy TSP algorithm, and alter the results by reversing segments."
    return alter_tour(greedy_tsp(cities))
```

```python
def alter_tour(tour):
    "Try to alter tour for the better by reversing segments."
    original_length = tour_length(tour)
    for (start, end) in all_segments(len(tour)):
        reverse_segment_if_better(tour, start, end)
    # If we made an improvement, then try again; else stop and return tour.
    if tour_length(tour) < original_length:
        return alter_tour(tour)
    return tour

def all_segments(N):
    "Return (start, end) pairs of indexes that form segments of tour of length N."
    return [(start, start + length)
            for length in range(N, 2-1, -1)
            for start in range(N - length + 1)]
```

```python
def reverse_segment_if_better(tour, i, j):
    "If reversing tour[i:j] would make the tour shorter, then do it." 
    # Given tour [...A-B...C-D...], consider reversing B...C to get [...A-C...B-D...]
    A, B, C, D = tour[i-1], tour[i], tour[j-1], tour[j % len(tour)]
    # Are old edges (AB + CD) longer than new ones (AC + BD)? If so, reverse segment.
    if distance(A, B) + distance(C, D) > distance(A, C) + distance(B, D):
        tour[i:j] = reversed(tour[i:j])
```

```python
plot_tsp(altered_greedy_tsp, cities)
```

Algortihm | Time Complexity | Length | Computation Time
--- | --- | --- | --- |
Greedy Algorithm | O(n<sup>2</sup>logn) | 1626.0 | 0.123 sec
Altered Greedy Algorithm | O(n<sup>2</sup>logn) | 1411.8 | 1.373 sec 

Altered Greedy Algorithm generated a shorter tour, but in a higher computation time (1.373 sec).

## Visualize greedy tsp segments

It's really helpful to visualise the process of segments' creation.   

```python
def visualize_greedy_tsp(cities, plot_sizes):
    """Go through edges, shortest first. Use edge to join segments if possible.
    Plot segments at specified sizes."""
    edges = shortest_edges_first(cities) # A list of (A, B) pairs
    endpoints = {c: [c] for c in cities} # A dict of {endpoint: segment}
    
    for (A, B) in edges:
        
        if A in endpoints and B in endpoints and endpoints[A] != endpoints[B]:
            new_segment = join_endpoints(endpoints, A, B)
            
            plot_segments(endpoints, plot_sizes, distance(A, B))
            if len(new_segment) == len(cities):
                return new_segment
        
def plot_segments(endpoints, plot_sizes, dist):
    "If the number of distinct segments is one of plot_sizes, then plot segments."
    segments = set(map(tuple, endpoints.values()))
    
    if len(segments) in plot_sizes:
        for s in segments:
            plot_lines(s)
        
        plt.show()
        plt.figure(figsize=(8, 8))
        print('{} segments, longest edge = {:.0f}'.format(
              len(segments), dist))
```

```python
visualize_greedy_tsp(cities, (50, 25, 10, 5, 2, 1));
```

We notice that the Greedy algorithm starts with segments of 1 city and constructs slowly bigger segments.
